```python
from pydantic import BaseModel, Field
from typing import List


class ComplexList(BaseModel):
    values: List[complex] = Field(default_factory=list)
```

```python
ComplexList()
```

```python
list_of_complexes = ComplexList(values=[1 + 2j, 3 + 4j])
list_of_complexes
```

```python
json = list_of_complexes.model_dump_json(indent=2)
print(json)
```

```python
ComplexList.model_validate_json(json)
```

```python
list_of_reals = ComplexList(values=[1.5, 2.5])
list_of_reals
```

```python
ComplexList.model_validate_json('{ "values": ["1.5+0j","2.5+0j" ]}')
```

```python
ComplexList.model_validate_json('{ "values": ["1.5","2.5+0j" ]}')
```

```python
ComplexList.model_validate_json('{ "values": [1.5,"2.5+0j" ]}')
```
